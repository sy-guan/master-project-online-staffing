```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read data
df = pd.read_csv('coople_data.csv')
```

```python
# Hourly salary histogram
df["Pay per hour"].plot(kind = 'hist')
print(df["Pay per hour"].mean())
```

```python
# Total salary histogram
df["Pay total"].plot(kind = 'hist')
print(df["Pay total"].mean())
```

```python
# Industry and frequency
d = {}
for i in df["Industry"]:
    if i not in d.keys():
        d[i] = 1
    else:
        d[i] += 1
d.pop(np.nan)
print(d)
```

```python
# Industry category
industry_cat = d.keys()
industry_cat
```

```python
# Industry bar plot
names = list(d.keys())
values = list(d.values())

plt.bar(range(len(d)), values, tick_label=names)
plt.show()
```

```python
# Histogram with industry and salary, return mean
def plot_hist_industry(industry,salary):
    l = []
    for index, row in df.iterrows():
        if row['Industry'] == industry:
            l.append(row[salary])
    plt.hist(l, bins=10, edgecolor='black')
    plt.xlabel(salary)
    plt.ylabel('Frequency')
    plt.show()
    return sum(l)/len(l)
```

```python
# Hotel hourly salary histogram
hotel = plot_hist_industry('Gastronomie / Hotel','Pay per hour')
hotel
```
